## Tensor Powers and Characters

Let $V$ be a $\mathbb{C}[G]$-module of dimension $n$. Let $\chi$ be the character for the action of $G$ on $V$. Fix $g \in G$. Let $e_1, \dots, e_n$ (resp. $\lambda_1, \dots, \lambda_n$) be the eigenvectors (resp. eigenvalues) for the action of $g$ on $V$. So $\chi(g) = \lambda_1 + \dots + \lambda_n$.

### Question 1
**What is the dimension of $V^{\otimes 3} = V \otimes V \otimes V$?**

**Answer:**
The dimension is $n \cdot n \cdot n = \mathbf{n^3}$.

### Question 2
**What are the eigenvectors (resp. eigenvalues) for the action of $g$ on $V^{\otimes 3}$?**

**Answer:**
A basis for $V^{\otimes 3}$ is given by the tensors $\{e_i \otimes e_j \otimes e_k\}$ for $1 \le i, j, k \le n$. 
Since $g \cdot e_i = \lambda_i e_i$, the action on the tensor is:
$$g(e_i \otimes e_j \otimes e_k) = (g e_i) \otimes (g e_j) \otimes (g e_k) = (\lambda_i e_i) \otimes (\lambda_j e_j) \otimes (\lambda_k e_k) = (\lambda_i \lambda_j \lambda_k) (e_i \otimes e_j \otimes e_k)$$
Thus, the **eigenvectors** are $\{e_i \otimes e_j \otimes e_k\}$ and the **eigenvalues** are $\mathbf{\lambda_i \lambda_j \lambda_k}$.

### Question 3
**Find constants $c_1$ and $c_2$ such that $n^3 = c_1 \binom{n}{3} + c_2 n(n - 1) + n$.**

**Answer:**
We partition the set of index triples $(i, j, k)$ based on their overlaps:
1. **Distinct indices** ($i, j, k$ are all different): There are $\binom{n}{3}$ sets of 3 distinct indices. Each set generates $3! = 6$ triples. So $c_1 = \mathbf{6}$.
2. **Two distinct indices** ($i$ appears once, $j$ appears twice, or vice versa): There are $\binom{n}{2}$ ways to pick the two values, and for each pair, there are 2 choices for which value repeats. For each such choice (e.g., $\{i, i, j\}$), there are $\frac{3!}{2!1!} = 3$ permutations. Total: $2 \binom{n}{2} \cdot 3 = n(n-1) \cdot 3$. So $c_2 = \mathbf{3}$.
3. **One distinct index** ($i=j=k$): There are $n$ such triples.

Verification: $6 \frac{n(n-1)(n-2)}{6} + 3n(n-1) + n = (n^3 - 3n^2 + 2n) + (3n^2 - 3n) + n = n^3$.

### Question 4

These are the multinomial coefficients for $( \sum \lambda_i )^3$:

### Question 6
**For $\Lambda^3(V)$ (the third exterior power of $V$):**

(a) **What is the dimension?**
The dimension is $\binom{n}{3} = \frac{n(n-1)(n-2)}{6}$.

(b) **Find a basis in terms of $e_1, \dots, e_n$.**
A basis is given by $\{e_i \wedge e_j \wedge e_k\}$ for $1 \le i < j < k \le n$.

(c) **What is the eigenvalue for the action of $g$ on the basis element?**
The action is $g(e_i \wedge e_j \wedge e_k) = (\lambda_i \lambda_j \lambda_k) (e_i \wedge e_j \wedge e_k)$. The eigenvalue is $\mathbf{\lambda_i \lambda_j \lambda_k}$.

(d) **Express the character for $\Lambda^3(V)$ as a linear combination of $A, B, C$.**
$\chi_{\Lambda^3(V)}(g) = \sum_{i<j<k} \lambda_i \lambda_j \lambda_k = \mathbf{A}$.

(e) **Express $\chi_{\Lambda^3(V)}(g)$ in terms of $\chi(g)^3, \chi(g)\chi(g^2)$, and $\chi(g^3)$.**
From the identities $A = \frac{1}{6} ( (6A+3B+C) - 3(B+C) + 2C )$, we have:
$$\chi_{\Lambda^3(V)}(g) = \frac{1}{6}\left( \chi(g)^3 - 3\chi(g)\chi(g^2) + 2\chi(g^3) \right)$$

### Question 7
**For $\text{Sym}^3(V)$ (the third symmetric power of $V$):**

(a) **What is the dimension?**
The dimension is $n$ multichoose 3, denoted $\left(\!\binom{n}{3}\!\right)$. Using the "Stars and Bars" formula (arranging 3 stars among $n$ bins using $n-1$ bars), this is $\binom{n+3-1}{3} = \binom{n+2}{3} = \frac{n(n+1)(n+2)}{6}$.

(b) **Find a basis in terms of $e_1, \dots, e_n$.**
A basis is given by the symmetric products $\{e_i e_j e_k\}$ (or $e_i \odot e_j \odot e_k$) for $1 \le i \le j \le k \le n$.

(c) **What is the eigenvalue for the action of $g$ on the basis element?**
The eigenvalue is $\mathbf{\lambda_i \lambda_j \lambda_k}$.

(d) **Express the character for $\text{Sym}^3(V)$ as a linear combination of $A, B, C$.**
$$\chi_{\text{Sym}^3(V)}(g) = \sum_{1 \le i \le j \le k \le n} \lambda_i \lambda_j \lambda_k = \mathbf{A + B + C}$$

(e) **Express $\chi_{\text{Sym}^3(V)}(g)$ in terms of $\chi(g)^3, \chi(g)\chi(g^2)$, and $\chi(g^3)$.**
From the identities $A + B + C = \frac{1}{6} ( (6A+3B+C) + 3(B+C) + 2C )$, we have:
$$\chi_{\text{Sym}^3(V)}(g) = \frac{1}{6}\left( \chi(g)^3 + 3\chi(g)\chi(g^2) + 2\chi(g^3) \right)$$

```python
# Sage Verification (Extended)
def verify_identities(n):
    print(f"Verifying for dimension n = {n}...")
    L = [var(f'L_{i}') for i in range(1, n+1)]
    
    A = sum(L[i]*L[j]*L[k] for i in range(n) for j in range(i+1, n) for k in range(j+1, n))
    B = sum(L[i] * (L[j]**2) for i in range(n) for j in range(n) if i != j)
    C = sum(l**3 for l in L)
    
    chi_g = sum(L)
    chi_g2 = sum(l**2 for l in L)
    chi_g3 = sum(l**3 for l in L)
    
    chi_wedge3 = (1/6) * (chi_g^3 - 3*chi_g*chi_g2 + 2*chi_g3)
    chi_sym3 = (1/6) * (chi_g^3 + 3*chi_g*chi_g2 + 2*chi_g3)
    
    print("Identity A == chi_Lambda3(V)?", bool(A == chi_wedge3))
    print("Identity A+B+C == chi_Sym3(V)?", bool(A + B + C == chi_sym3))

verify_identities(3)
```

### Part 3: The Case $d = 5, k = 3$

Recall the isomorphism $\Lambda^k V \cong \Lambda^{n-k} V \otimes \text{sgn}$, or specifically for the standard representation of $S_d$ (where $n = \dim V = d-1$): $\Lambda^k V \cong \Lambda^{d-1-k} V \otimes \text{sgn}$.

**a) Show that for $S_5$, $\Lambda^3 V \cong \Lambda^1 V \otimes \text{sgn}$.**

**Answer:**
For $S_5$, the degree is $d = 5$. The standard representation $V$ has dimension $n = d - 1 = 4$. Using $k = 3$, the general isomorphism gives:
$$\Lambda^3 V \cong \Lambda^{4-3} V \otimes \text{sgn} = \Lambda^1 V \otimes \text{sgn}.$$

**b) Since $V$ is irreducible and the sign representation is 1-dimensional, why does this immediately imply $\Lambda^3 V$ is irreducible?**

**Answer:**
The tensor product of any irreducible representation with a 1-dimensional representation is always irreducible. Since $V = \Lambda^1 V$ is irreducible, $V \otimes \text{sgn}$ is irreducible. By the isomorphism from part (a), $\Lambda^3 V$ must also be irreducible.

**c) Draw the Young diagram for the partition $(2, 1, 1, 1)$. Verify that this is the conjugate partition to the one representing the standard representation of $V$ of dimension 4.**

**Answer:**
The partition $(2, 1, 1, 1)$ is a hook with 5 boxes:
$$\begin{array}{l}
\Box \; \Box \\
\Box \\
\Box \\
\Box
\end{array}$$

The standard representation $V$ of $S_5$ corresponds to the partition $(4, 1)$:
$$\begin{array}{l}
\Box \; \Box \; \Box \; \Box \\
\Box
\end{array}$$
Conjugating (transposing) the Young diagram for $(4, 1)$ swaps rows and columns, resulting exactly in the diagram for $(2, 1, 1, 1)$. Tensor multiplication by the sign representation maps an irreducible representation to the one corresponding to its conjugate partition. Hence $\Lambda^3 V \cong V \otimes \text{sgn}$ corresponds to $(2, 1, 1, 1)$.

### Discussion Questions

**How does the branching rule apply to $\Lambda^k V$ when restricting from $S_d$ to $S_{d-1}$?**

**Answer:**
The representation $\Lambda^k V$ corresponds to the hook partition $(d-k, 1^k)$. By the branching rule, the restriction to $S_{d-1}$ decomposes into a direct sum of representations obtained by removing one box from the Young diagram in all valid ways. There are exactly two removable boxes (the ends of the hook): 
1. Removing from the first row gives $(d-1-k, 1^k)$, which corresponds to $\Lambda^k V'$.
2. Removing from the first column gives $(d-k, 1^{k-1})$, which corresponds to $\Lambda^{k-1} V'$.
Thus, $\text{Res}_{S_{d-1}}^{S_d}(\Lambda^k V) \cong \Lambda^k V' \oplus \Lambda^{k-1} V'$, where $V'$ is the standard representation of $S_{d-1}$. This mirrors Pascal's triangle identity for binomial coefficients: $\binom{n}{k} = \binom{n-1}{k} + \binom{n-1}{k-1}$.

**Fulton & Harris use the fact that $V$ is the "Stiefel" representation. How do the weights of the action of the diagonal subgroup in $GL_d(\mathbb{C})$ help identify these specific representations as hooks?**

**Answer:**
In the context of the symmetric group embedded in $GL_d(\mathbb{C})$, the standard basis vectors $e_1, \dots, e_d$ of $\mathbb{C}^d$ are weight vectors for the diagonal torus, with weights $L_1, \dots, L_d$. The exterior powers $\Lambda^k(\mathbb{C}^d)$ have bases given by $e_{i_1} \wedge \dots \wedge e_{i_k}$, which correspond to the weights $L_{i_1} + \dots + L_{i_k}$. The highest weight of $\Lambda^k(\mathbb{C}^d)$ is $L_1 + \dots + L_k$, which precisely corresponds to the fundamental weight $\omega_k$, represented by a vertical column of $k$ boxes. By tracking how $\mathbb{C}^d = V \oplus \text{trivial}$ restricts, the characters of exterior powers $\Lambda^k V$ split into forms corresponding exactly to hook-shaped Young diagrams.
