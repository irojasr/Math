```python
### Worksheet time 20260325 ###

### 1. Define G and H as a subgroup of G

# 1. Define the generators
# x_cycle is the generator for C7 (the normal subgroup)
x_cycle = '(1,2,3,4,5,6,7)'

# y_action is the generator for C3 acting on C7 via x -> x^2
# # In cycle notation, x -> 2x mod 7 corresponds to (1,2,4)(3,6,5)
# y_action = '(1,2,4)(3,6,5)'

# The inverse of (1,2,4)(3,6,5) is (1,4,2)(3,5,6)
y_action = '(1,4,2)(3,5,6)'

# 2. Define the group as a PermutationGroup
G = PermutationGroup([x_cycle, y_action])

# 3. Define C7 as a subgroup
C7_subgroup = G.subgroup([x_cycle])

# --- Verification ---
print(f"Group Order: {G.order()}")      # Result: 21
print(f"Is Abelian: {G.is_abelian()}")  # Result: False
print(f"Subgroup Order: {C7_subgroup.order()}") # Result: 7

# Define x and y explicitly using the group G
x = G(x_cycle)
y = G(y_action)
# Now check the relations
print(f"Is N normal? {(y * x * y**-1) in G.subgroup([x])}") 
print(f"yxy^-1 == x^2: {y * x * y**-1 == x**2}")
print(f"yxy^-1 == x^4: {y * x * y**-1 == x**4}") # The other non-trivial possibility

# Find 'a' such that yxy^-1 = x^a
result = y * x * y**-1
for a in range(1, 7):
    if result == x**a:
        print(f"y acts as x^{a}!")

### 2. Find the character tables of G and H

# Compute the irreducible characters for the semidirect product G (order 21)
# These represent the different ways G can act on vector spaces
G_irreps = G.irreducible_characters()
# RESTRICT each character of G to the C7 subgroup
# For each character chi of G, this computes the value of chi only on the C7 part.
# The result 'chi1C7' is a list of characters of the subgroup C7
chi1C7 = [x.restrict(C7_subgroup) for x in G_irreps]
# PRINT the character table of G
# Each printed row corresponds to an irreducible character of G
# The values are its values on each of the conjugacy classes of G
print("--- Character Table of G (C7 \rtimes C_3) ---")
for k in range(len(G_irreps)):
    print(G_irreps[k].values())
# PRINT the restricted characters of G to C7
# Each row shows how an irreducible representation of G "breaks down" or behaves
# when you restrict its action only to the C7 subgroup.
print("\n--- Characters of G restricted to C7 ---")
for k in range(len(chi1C7)):
    print(chi1C7[k].values())
# To get the ACTUAL Character Table of C7 itself (order 7):
# You can compare these restricted values against the pure character table of C7.
print("\n--- Irreducible Characters of C7 ---")
for chi in C7_subgroup.irreducible_characters():
    print(chi.values())
```
