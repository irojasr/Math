```python
### Comprehensive Visualization of Congruence Subgroups N=2 to 8 ###

# Point to map: tau = 1/2
pt = 1/2

# Hardcode basic recognizable color strings
colors = ["red", "blue", "green", "purple", "orange", "cyan", "magenta", "maroon", "olive", "navy", "teal", "yellow"]

for n in range(2, 9):
    print("\n" + "="*20 + f" LEVEL N={n} " + "="*20)
    
    # Iterate progressively through the subgroups in exact requested order: Gamma, Gamma0, Gamma1
    for name, func in [("Gamma", Gamma), ("Gamma0", Gamma0), ("Gamma1", Gamma1)]:
        G = func(n)
        idx = G.index()
        
        print(f"\n--- Investigating {name}({n}) --- (Index in SL_2(Z) = {idx})")
        
        # Use a try block because FareySymbol doesn't always natively support raw Gamma(N) geometries 
        try:
            # We enforce using the explicit coset geometric transversal tied to the Farey domain renderer
            reps = list(FareySymbol(G).coset_reps())
            P = FareySymbol(G).fundamental_domain().plot()
            
            # Avoid crashing the browser by limiting raw console text output if there are too many matrices
            if len(reps) <= 30:
                print("Mapping Results:")
            else:
                print(f"Mapping Results: [{len(reps)} mapped points generating geometry secretly. Output muted due to huge index!]")
                
            for i, M in enumerate(reps):
                a, b, c, d = M[0,0], M[0,1], M[1,0], M[1,1]
                
                # Apply the Moebius transformation via the mathematical *inverse* matrix over pt = 1/2!
                mapped_pt = (d*pt - b)/(-c*pt + a)
                
                c_name = colors[i % len(colors)]
                P += point((mapped_pt.real(), mapped_pt.imag()), color=c_name, size=80, zorder=10)
                
                if len(reps) <= 30:
                    print(f"Matrix: [[{a}, {b}], [{c}, {d}]] - Image of 1/2: {mapped_pt} - Color: {c_name}")
            
            # Widen the horizontal layout safely!
            if name == "Gamma":
                # Standard aspect for the main principal loop
                P.show(gridlines=True, title=f"Action on 1/2 for {name}({n})", 
                       xmin=-0.1, xmax=n+0.1, ymin=0, ymax=2.2)
            else:
                # Lowering ymax and forcing 'automatic' aspect ratio entirely unlocks the plot's width!
                P.show(gridlines=True, title=f"Action on 1/2 for {name}({n})", 
                        xmin=-0.1, xmax=1+0.1, ymin=0, ymax=2.2)
                    #    xmin=0, xmax=1, ymin=0, ymax=4, 
                    #    aspect_ratio='automatic', figsize=[10, 7.5])
            #, figsize=[15, 5] up here
        except Exception as e:
            print(f"[!] Visualization skipped due to mathematical error or missing support in FareySymbol: {e}")

print("\n" + "="*50)
```
