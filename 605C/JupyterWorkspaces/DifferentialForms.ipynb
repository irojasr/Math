# Differential Forms on Elliptic Curves
## The Holomorphicity of $\omega = dx/y$ at Infinity


Consider the elliptic curve $E$ given by the affine equation:
$$y^2 = x^3 - x$$

We want to show that the differential form $\omega = \frac{dx}{y}$ has no poles at the point at infinity. To do this, we transform the form into the $(u, w)$ chart (projective closure at $Y=1$).


### 1. Homogenization
The projective equation $F(X, Y, Z) = 0$ is obtained by substituting $x = X/Z$ and $y = Y/Z$:
$$(Y/Z)^2 = (X/Z)^3 - (X/Z) \implies Y^2 Z = X^3 - XZ^2$$

### 2. Change of Chart (Dehomogenizing at $Y=1$)
To rigorously study the point at infinity $(0:1:0)$, we look deeply at the local affine chart where $Y=1$. 
Let $u = X/Y$ and $w = Z/Y$. The point at infinity corresponds exactly to the origin $(u,w) = (0,0)$.

The original affine coordinates $(x, y)$ map relationally to $(u, w)$ as:
$$x = \frac{X}{Z} = \frac{X/Y}{Z/Y} = \frac{u}{w}$$
$$y = \frac{Y}{Z} = \frac{1}{w}$$

### 3. Substituting into $\omega$
We calculate $dx$ strictly using generalized differentials:
$$dx = d\left(\frac{u}{w}\right) = \frac{w\,du - u\,dw}{w^2}$$

Substituting $dx$ and $y$ backwards against $\omega$:
$$\omega = \frac{dx}{y} = \frac{\frac{w\,du - u\,dw}{w^2}}{\frac{1}{w}} = \frac{w\,du - u\,dw}{w} = du - \frac{u}{w}dw$$


### 4. Relating $du$ and $dw$ via the Curve Equation
In our local $Y=1$ chart, the bounding curve equation $Y^2 Z = X^3 - XZ^2$ forces the coordinate constraint:
$$w = u^3 - uw^2$$

Differentiating both sides natively:
$$dw = 3u^2\,du - (w^2\,du + 2uw\,dw)$$
$$\implies dw(1 + 2uw) = (3u^2 - w^2)du$$
$$\implies dw = \frac{3u^2 - w^2}{1 + 2uw}du$$

We then formally substitute $dw$ straight back into our earlier expression for $\omega$:
$$\omega = du - \frac{u}{w} \left( \frac{3u^2 - w^2}{1 + 2uw} \right) du = \left( 1 - \frac{3u^3 - uw^2}{w(1 + 2uw)} \right) du$$

### 5. Algebraic Reduction at the Pole Boundary
We can eliminate the $u^3$ numerator term entirely by plugging our curve constraint ($u^3 = w + uw^2$) back in:
$$\omega = \left( 1 - \frac{3(w + uw^2) - uw^2}{w(1 + 2uw)} \right) du$$
$$\omega = \left( 1 - \frac{3w + 2uw^2}{w(1 + 2uw)} \right) du$$
$$\omega = \left( 1 - \frac{w(3 + 2uw)}{w(1 + 2uw)} \right) du$$
$$\omega = \left( 1 - \frac{3 + 2uw}{1 + 2uw} \right) du$$

Algebraically evaluating terms directly resolves the limit boundary:
$$\omega = \left( \frac{1 + 2uw - 3 - 2uw}{1 + 2uw} \right) du = \frac{-2}{1 + 2uw} du$$

Evaluating identically at the point at infinity $(u, w) = (0, 0)$:
$$\omega(0,0) = \frac{-2}{1} du = -2 du$$
Because the expansion coefficient evaluates precisely to a constant structural nonzero integer ($-2$), the boundary forms identically uniformly. **There are conclusively zero poles at infinity!**

Let $X$ be a smooth, irreducible projective curve (or a compact Riemann Surface). Let $\omega \in \mathcal{K}(X)^1$ be a non-zero meromorphic 1-form.

### The Divisor of a 1-Form
In a local coordinate chart $(U, z)$ around a point $P \in X$, we can write $\omega = f(z)dz$ for some meromorphic function $f$. The order of $\omega$ at $P$ is defined as:
$$v_P(\omega) = v_P(f)$$

The **canonical divisor** associated to $\omega$ is the formal sum:
$$(\omega) = \sum_{P \in X} v_P(\omega) \cdot P$$
All such divisors are linearly equivalent, forming the **canonical class** $K_X$.


### Pullback and the Hurwitz Formula
Let $\phi: X \to Y$ be a non-constant holomorphic map of degree $d$ between smooth curves. For a 1-form $\eta$ on $Y$, the pullback $\phi^* \eta$ is a 1-form on $X$.

If $P \in X$ and $Q = \phi(P) \in Y$, let $e_P$ be the ramification index at $P$. If $y$ is a local coordinate at $Q$ and $z$ is a local coordinate at $P$ such that $\phi(z) = z^{e_P} = y$, then:
$$\phi^*(dy) = d(z^{e_P}) = e_P z^{e_P - 1} dz$$


#### Example 1: $\mathbb{P}^1$ and the form $dx$
For $X = \mathbb{P}^1$, $f = x$, then $div(x) = 1[0] - 1[\infty]$. Thus, $div(dx) = -2[\infty]$.

#### Example 2: Elliptic Curve Pullback Table
For an elliptic curve $E: y^2 = x^3 - x$, we have the map $\phi: E \to \mathbb{P}^1$ which comes from the pullback of the map $\mathbb{C}[x] \to \mathbb{C}[x,y]/(y^2 - x^3 - x)$. There is ramification above $0, 1, -1$ and $\infty$.

Using $\eta = dx$ on $\mathbb{P}^1$, let's trace the zeros and poles across the pullback:

| Type | $v_Q(dx)$ (Zeros or Poles) | Ramification Index $e_P$ | $v_P(\phi^*(dx))$ (Zeros or Poles) | Formula: $e_P \cdot v_Q(\omega) + (e_P - 1)$ |
|---|---|---|---|---|
| **Regular points** | 0 | 1 | 0 | $1(0) + (1-1) = 0$ |
| **Branch points $\{0, 1, -1\}$** | 0 | 2 | 1 (Simple Zero) | $2(0) + (2-1) = 1$ |
| **Infinity $\infty$** | -2 (Pole of order 2) | 2 | -3 (Pole of order 3) | $2(-2) + (2-1) = -3$ |

From here, we explicitly calculate the divisor:
$$div(dx) = P_{-1} + P_0 + P_1 - 3P_{\infty}$$

If we analyze the poles and zeros of the uniformizer $y$ natively, $y$ has simple zeros at the three finite branch points and a pole of order 3 at $\infty$. So $div(y) = P_{-1} + P_0 + P_1 - 3P_{\infty}$.

Thus, $div(dx/y) = div(dx) - div(y) = 0$. The differential is holomorphic!


### Formula for the Divisor of a Pullback
The order of the pullback $\phi^* \eta$ at $P$ is formally given as:
$$v_P(\phi^* \eta) = e_P \cdot v_{\phi(P)}(\eta) + (e_P - 1)$$
Equivalently, in terms of divisors: $$(\phi^*(\eta)) = \phi^*(\eta) + R$$
Where $R = \sum_{P \in X} (e_P - 1) P$ is exactly the **ramification divisor**.

### Addendum: Proof that $\deg(K_X) = 2g - 2$ via Riemann-Roch

The Riemann-Roch theorem for a divisor $D$ on a smooth projective curve $X$ of genus $g$ states:
$$\ell(D) - \ell(K_X - D) = \deg(D) - g + 1$$

To find the degree of the canonical divisor $K_X$, we simply let $D = K_X$. The formula becomes:
$$\ell(K_X) - \ell(0) = \deg(K_X) - g + 1$$

Now, we evaluate the two dimensions on the left hand side:
1. **$\ell(K_X)$**: This is the dimension of the vector space of global holomorphic 1-forms (Abelian differentials) on $X$. By definition, precisely this dimension is the genus $g$, so $\ell(K_X) = g$.
2. **$\ell(0)$**: This is the dimension of the space of global holomorphic functions. On a compact Riemann surface, by the Maximum Modulus Principle, the only globally holomorphic meromorphic functions are constants. Thus, the space is simply $\mathbb{C}$, meaning $\ell(0) = 1$.

Substituting these natively into our evaluated formula:
$$g - 1 = \deg(K_X) - g + 1$$

Solving mapping properties cleanly explicitly for $\deg(K_X)$, we elegantly seamlessly shift the constants:
$$\deg(K_X) = g + g - 1 - 1 = 2g - 2$$

Thus, the canonical divisor mathematically evaluating globally must identically firmly bounds to precisely degree $2g-2$!

### The Space of Canonical Divisors
Recall that $\text{Div}(X) = \sum_{P\in X} a_P [P]$ is the free Abelian group on classes $[P]$. Inside this group, we have the degree zero divisors (those which sum to zero), and principal divisors inside there which are all of the form $\text{div}(f)$.

However, inside $\text{Div}(X)$ we also have $\text{Div}^{2g-2}(X)$ (as a subset) consisting of divisors of degree $2g-2$. Further inside, we have the canonical divisors $KDiv(X)$, which are all divisors of one-forms.

We would like to understand what the space of canonical divisors looks like. In fact, we have two key properties:
1. $\text{div}(f\omega) = \text{div}(f) + \text{div}(\omega)$ for a one-form $\omega$.
2. $KDiv(X)$ is the coset of the principal divisors in $\text{Div}(X)$ containing $\text{div}(\omega)$.


To prove this, say we take a principal divisor $D = \text{div}(f)$. We wish to show that:
$$\text{div}(\omega) + D \in KDiv(X)$$
By our first property, $\text{div}(\omega) + \text{div}(f) = \text{div}(f\omega)$, which is exactly the divisor of the one-form $f\omega$. This shows that the entire coset of principal divisors is contained within the canonical divisors.

On the other side, if we take another canonical divisor $D_2 = \text{div}(\omega_2)$, we wish to show that their difference locally looks like:
$$\text{div}(g_1 dz) - \text{div}(g_2 dz)$$

If we let $f = g_1 / g_2$, then $\text{div}(f) = \text{div}(g_1) - \text{div}(g_2)$. This also immediately implies that every canonical divisor must have exactly the same degree (which we proved previously is $2g-2$).


In essence, taking two canonical divisors $D_1 = \text{div}(\omega_1)$ and $D_2 = \text{div}(\omega_2)$, we show they are in the exact same coset by looking at their difference:
$$D_1 - D_2 = \text{div}(\omega_1) - \text{div}(\omega_2) = \sum_{P} (\text{ord}_P(g_1(z)) - \text{ord}_P(g_2(z)))[P] = \sum_{P} \text{ord}_P(g_1/g_2)[P] = \text{div}(g_1/g_2)$$

Because their difference resolves perfectly to a principal divisor $\text{div}(f)$, we can decisively conclude that $KDiv(X)$ is exactly a single coset of the principal divisors!


### Pullbacks of One-Forms (Revisited)
If $F: X \to Y$ is a holomorphic map, we can pull back one-forms from $Y$ to $X$. The pullback acts on functions via $(F^*f)(w) = f(g(w))$, and for one-forms:
$$F^*(\omega) = f(g(w)) g'(w) dw$$

For functions, the order of vanishing multiplies by the ramification index: $\text{ord}_P(F^*f) = n \cdot k$, where $k = \text{ord}_{F(P)}(f)$ and $n = \text{mult}_P(F)$ is the ramification index.

For one-forms, the formula completes to:
$$\text{ord}_P(F^*\omega) = n \cdot \text{ord}_{F(P)}(\omega) + (n - 1)$$

**Proof:** Near $P$, we may locally write $z = w^n$. If $\omega = (c z^k + O(z^{k+1}))dz$, then:
$$F^* \omega = (c (w^n)^k + O(w^{nk+1}))(n w^{n-1} dw) = (n c w^{nk + n - 1} + O(w^{nk+n})) dw$$
The new order is strictly $nk + n - 1$, finalizing exactly what was missing.


### Existence of Canonical Divisors
**Theorem:** Let $X$ be a compact Riemann Surface of genus $g$ with meromorphic functions $\mathcal{M}(X) \neq \mathbb{C}$. Then there is a canonical divisor on $X$ with degree $2g-2$.
*(This theorem basically says: if you believe you have a function on your curve, you guaranteed have a one-form.)*

**Proof:** By hypothesis, there is a non-constant function $F: X \to \mathbb{P}^1$. Let $\eta = F^*(dz)$. On $\mathbb{P}^1$, $\text{div}(dz) = -2[\infty]$. We just need to check the pullback degree:

$$\deg(\text{div}(\eta)) = \sum_{P \in X} \text{ord}_P(F^* dz)$$
$$= \sum_{F(P) \neq \infty} (\text{mult}_P(F) - 1) + \sum_{F(P) = \infty} (\text{mult}_P(F)(-2) + \text{mult}_P(F) - 1)$$
$$= \sum_{F(P) \neq \infty} (\text{mult}_P(F) - 1) + \sum_{F(P) = \infty} (-\text{mult}_P(F) - 1)$$
$$= \left( \sum_{P \in X} (\text{mult}_P(F) - 1) \right) - 2 \left( \sum_{F(P) = \infty} \text{mult}_P(F) \right)$$

Notice that the first sum evaluates directly to the Riemann-Hurwitz Ramification index $R = \sum (\text{mult}_P(F)-1)$, and the second sum is the total degree of the map $2\deg(F)$.
$$= \deg(R) - 2\deg(F)$$

We know from the Riemann-Hurwitz equation: 
$$2g - 2 = d(-2) + \deg(R) \implies -2\deg(F) + \deg(R) = 2g - 2$$
Thus: 
$$= 2g - 2$$
